```python
numbers = [0, 1, 2, 3, 4]

tuples = [(i, j, k) for i in range(len(numbers)-2) for j in range(i+1, len(numbers)-1) for k in range(j+1, len(numbers))]
for t in tuples:
    print(t)
```

```python
import timeit
l = [100, 200, 300]

def comprehension(n):
    start = timeit.default_timer()
    sets = {(i, j) for i in range(n-1) for j in range(i+1, n)}
    end = timeit.default_timer()
    return (sets, end-start)

for i in l:
    result, time_taken = comprehension(i)
    print(f'Time taken: {time_taken:.6f} seconds')
    print(result)
```

```python
def unions(n):
    start = timeit.default_timer()
    pairSet = set()

    for i in range(n-1):
        for j in range(i+1, n):
            pairSet = pairSet.union({(i, j)})
    end = timeit.default_timer()
    return (pairSet, end-start)


for i in l:
    result, time_taken = unions(i)
    print(f'Time taken: {time_taken:.6f} seconds')
    print(result)
```
