```python
path = 'PoliticalContributions/committee_master.txt'
nameDict = {}

with open(path) as f:
    for line in f:
        data = line.split('|')
        if data[1] != '':
            nameDict[data[0]] = data[1]
```

```python
print(f'Number of committees: {len(nameDict)}')
committees = set(nameDict.keys())
```

```python
path = 'PoliticalContributions/itoth.txt'
contributorDict = {}
with open(path) as f:
    for line in f:
        data = line.split('|')
        contributor = data[0]

        if contributorDict.get(contributor) is None:
            contributorDict[contributor] = {data[7]}
        else:
            contributorDict[contributor].add(data[7])
```

```python
n = len(contributorDict)
print(f'N pairs: {n*(n-1)/2}')
```

```python
for key in list(contributorDict.keys()):
    if len(contributorDict[key]) <= 500:
        contributorDict.pop(key, None)

n = len(contributorDict)
print(f'N pairs: {n*(n-1)/2}')
```

```python
for key in contributorDict:
    print(nameDict[key], len(contributorDict[key]))
```

```python
contributors = list(contributorDict.keys())
```

```python
pairs = [(contributors[i], contributors[j]) for i in range(n-1) for j in range(i+1, n, 1)]
```

```python
simDict = {}

for commA, commB in pairs:
    A = contributorDict[commA]
    nameA = nameDict[commA]
    B = contributorDict[commB]
    nameB = nameDict[commB]

    nIntersection = len(A.intersection(B))

    JaccardSimilarityAB = nIntersection/len(A.union(B))
    ProbabilityA_given_B = nIntersection/len(B)
    ProbabilityB_given_A = nIntersection/len(A)

    simDict[(nameA, nameB)] = (JaccardSimilarityAB, ProbabilityA_given_B, ProbabilityB_given_A)
```

```python
import operator

sortedList = sorted(simDict.items(), key=operator.itemgetter(1), reverse=True)
```

```python
for committees, simMeasures in sortedList:
    nameA, nameB = committees
    jAB, P_AgivenB, P_BgivenA = simMeasures

    if jAB > .5:
        print(f'--- Jaccard Similarity: {round(jAB, 3)} --- Probability of A given B: {round(P_AgivenB, 3)} --- Probability of B given A: {round(P_BgivenA, 3)} -------------- NAMES: {nameA} --- {nameB}')
```

```python
path = 'PoliticalContributions/committee_master.txt'
nameDict = {}

with open(path) as f:
    for line in f:
        data = line.split('|')
        if data[1] != '':
            nameDict[data[0]] = data[1]
```

```python
path = 'PoliticalContributions/itoth.txt'
contributionsToPACs = {}

with open(path) as f:
    for line in f:
        data = line.split('|')
        recipientCode = data[0]
        PACmakingThePaymentName = data[7]
        amount = data[14]

        recipientName = nameDict.get(recipientCode, recipientCode)

        if contributionsToPACs.get(recipientName) is None:
            contributionsToPACs[recipientName] = [(amount, PACmakingThePaymentName)]
        else:
            contributionsToPACs[recipientName].append((amount, PACmakingThePaymentName))
```

```python
import itertools

first_few = dict(itertools.islice(contributionsToPACs.items(), 10))

for k in first_few:
    print(f'------ Recipient: {k} ------ Transactions: {first_few[k]}')
```
