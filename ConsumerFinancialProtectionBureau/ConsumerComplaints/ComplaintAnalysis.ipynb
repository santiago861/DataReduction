```python
import pandas as pd

path = 'complaints.csv'
df = pd.read_csv(path)

df.head()
```

```python
target_issue = 'Loan modification,collection,foreclosure'

filtered_df = df[df['Issue'] == target_issue]
filtered_df
```

```python
produ_filter_df = filtered_df['Product'].unique()

print(f'The issue "{target_issue}" appears in the following product categories:')
for product in produ_filter_df:
    print(f'-- {product}')
```

```python
appear_counts_dict = {}

for row in filtered_df.itertuples():
    if appear_counts_dict.get(row.Company) is None:
        appear_counts_dict[row.Company] = 1
    else:
        appear_counts_dict[row.Company] += 1

print(appear_counts_dict)
```

```python
import operator

sorted_list = sorted(appear_counts_dict.items() , key=operator.itemgetter(1), reverse=True)
print('Company ----------------------- Counts')
for i in range(10):
    print(f'{sorted_list[i][0]} --- {sorted_list[i][1]}')
```
